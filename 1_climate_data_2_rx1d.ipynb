```python
import os
import glob
from datetime import datetime
from netCDF4 import Dataset
import netCDF4 as nc
from datetime import datetime
import pandas as pd
import numpy as np
import xarray as xr
import sys
sys.path.append('C:/Users/Administrator/anaconda3/envs/gdal/Lib/site-packages/')
sys.path.append('C:/Users/Administrator/anaconda3/envs/gdal/Lib/site-packages/lmoments3')
import distr
from scipy import stats
import matplotlib.pyplot as plt
import multiprocessing as mp
from tqdm import tqdm
import basefunc

def savenc(array,  lat, lon, time,save_path,nn='pr'):
    
    # 创建一个xarray.DataArray对象，这使得基于时间的操作更加容易
    da = xr.DataArray(array , coords=[time,lat, lon], dims=['time', 'lat', 'lon'],name=nn)
    # 重新采样数据为年度频率，并找到每年的最大值
    #save_path = os.path.join(path, 'MRI-ESM2-0(yearmax)/', ff+f'{star}_{end}.nc')
    da.to_netcdf(save_path)




''' Calculate the annual maximum daily precipitation '''
def calculate(file,scen,model): 
    
    save_folder = f'D:/QQ/PhD/cmip6_split/{model}_Rx1d/{scen}/'
    ref_tif='./data/ref.tif'
    for i in range(len(file)):
        file_name = os.path.basename(file[i])
        output_path=os.path.join(save_folder,file_name[:-3]+'.tif')
        if os.path.exists(output_path):
                continue
        else:
            
            rootgrp = Dataset(file[i])
            data=rootgrp.variables['pr'][:]
            lat=rootgrp.variables['lat'][:]
            lon=rootgrp.variables['lon'][:]
            array=data.max(0)[::-1,:] * 86400#*60*60*24 #convert to "mm"
            
            print(output_path)
            basefunc.array2Raster(array,ref_tif,output_path)
        
    

FILE_PATH = "D:\\QQ\\PhD\\cmip6_split\\{model}\\pr_day_{model}_{scenario}_r1i1p1f1_gn_{y}.nc" 

SCENARIO_YEARS = {
    'historical': [(1985, 2015)], 
    'ssp126': [(2030, 2060), (2070, 2100)],
    'ssp245': [(2030, 2060), (2070, 2100)],
    'ssp370': [(2030, 2060), (2070, 2100)],
    'ssp585': [(2030, 2060), (2070, 2100)],
}

def generate_files(scenario, model, years):
    files = []
    for y in years:
        files.append(FILE_PATH.format(model=model, scenario=scenario, y=y))
    return files
models = ['ACCESS-CM2','ACCESS-ESM1-5']

for scenario, ranges in SCENARIO_YEARS.items():
    for start, end in ranges:
        years = np.arange(start, end)
        for model in models:
            files = generate_files(scenario, model, years)
            calculate(files, scenario, model)
            print(f'done {scenario}')
```
